```python
import pandas as pd

# 读取CSV文件并指定换行符为'<br/>'，然后将'<br/>'替换为换行符'\n'
with open('reviews.csv', 'r', encoding='utf-8') as file:
    csv_content = file.read().replace('<br/>', '\n')


# 将替换后的内容写入新的临时文件，显式指定编码为'utf-8'
with open('temp_reviews.csv', 'w', encoding='utf-8') as file:
    file.write(csv_content)


# 使用 Pandas 读取替换后的临时文件
reviews = pd.read_csv('temp_reviews.csv')

# 删除临时文件
import os
os.remove('temp_reviews.csv')
```

```python
reviews['comments'] = reviews['comments'].str.replace('\n', '')
```

```python
reviews.shape
```

```python
reviews = reviews.dropna(subset=['comments'])
```

```python
from langdetect import detect, LangDetectException

# 假设df是你的评论数据DataFrame
def is_english(text):
    try:
        return detect(text) == 'en'
    except LangDetectException:
        return False

# 应用函数检查每条评论是否为英文
reviews['is_english'] = reviews['comments'].apply(is_english)

# 筛选出所有评论都是英文的listing_id
english_reviews = reviews[reviews['is_english']]
```
